## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

I am selecting the Content Performance Optimization and Decay Lane. The primary objective is to identify which high-impact web pages are experiencing traffic or ranking decay so they can be prioritized for content refreshes. This lane is critical because static content naturally loses relevance over time, and catching a decline early prevents severe traffic loss.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Decision: Which decaying or declining pages should our content team prioritize for a manual update or rewrite this week?

Action: The content team will pull the top-ranked pages from our system pipeline and manually refresh their text, update internal links, or optimize titles.

Cost of a wrong call: If we flag a page that isn't actually declining (a False Positive), we waste valuable editorial hours rewriting a healthy page. If we miss a page that is severely dying (a False Negative), we lose organic search traffic, resulting in a direct drop in views and revenue.

## 3. Quick look at the data (2-3 real numbers)

Based on the data execution below, this lane is justified by the following real numbers from the starter dataset:

Out of the 11,477 pages in our data, 3,251 pages (28.3%) are actively in a downward trend.

Declining pages exhibit a noticeably lower average click-through rate compared to healthy pages, demonstrating a clear loss of engagement that requires data-driven intervention.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What I can claim: This work will provide an observed, directional decision-support framework to flag pages showing statistical signs of traffic and ranking patterns associated with historical content decay.

What I cannot claim: I cannot claim to have cracked or predicted Google's core ranking algorithm, nor can I claim causal proof that our model's recommendations are the sole absolute cause of a page's organic performance shifts.

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Create the declining flag
df["is_declining"] = df["trend_direction"].str.lower().eq("down").astype(int)

# Calculate real numbers for our framing
total_pages = len(df)
declining_count = df["is_declining"].sum()
declining_pct = df["is_declining"].mean() * 100

print(f"Total pages in starter dataset: {total_pages}")
print(f"Number of declining pages: {declining_count}")
print(f"Percentage of pages currently declining: {declining_pct:.1f}%")
```
